## Exercise #3:
Write up a k-nearest-neighbors function like the one you made for the iris data set in class. It should be able to make mpg ("comb08") predictions for new cars based on the year, cylinders, and displacement. You should also be able to specify what you want to use as k.

Pay special attention to whether this problem is a classification or a regression, and how the answer to that question may change the output of your model. 

```python
## Build your function here
def knn(car, data, k):

  data['distance'] = numpy.sqrt(
    (car['year']-data['year'])**2
    +(car['cylinders']-data['cylinders'])**2
    +(car['displ']-data['displ'])**2)

  sorted_data = data.sort_values(['distance'])

  return sorted_data.head(k)['comb08'].mean()
```

## Exercise #4: 
Demonstrate that your function works by making up some new values for hypothetical cars and using your function to display the predicted mpg ("comb08") for that car. 

```python
## demonstrate that your function works here

car1 = {}
car1['year'] = 1985
car1['cylinders'] = 4
car1['displ'] = 5

car2 = {}
car2['year'] = 1993
car2['cylinders'] = 8
car2['displ'] = 2

print("First car mpg: ", knn(car1, subset, 5))
print("Second car mpg: ",knn(car2, subset, 3))
```

## Exercise #5
Make a copy of the data and normalize the training data using Z-score-- `copy()` will be useful here. Predict the mpg of a particular car using your k-nearest-neighbors function with both the normalized and non-normalized training data. Compare your results. Use a markup cell to describe and explain the differences in a few sentences.

```python
## normalize a copy of the data

# copy the data
norm_subset = subset.copy()

# means
year_mean = subset['year'].mean()
cylinders_mean = subset['cylinders'].mean()
displ_mean = subset['displ'].mean()

# stds
year_std = subset['year'].std()
cylinders_std = subset['cylinders'].std()
displ_std = subset['displ'].std()

# replacing the data with z scores
norm_subset['year'] = (norm_subset['year'] - year_mean)/year_std
norm_subset['cylinders'] = (norm_subset['cylinders'] - cylinders_mean)/cylinders_std
norm_subset['displ'] = (norm_subset['displ'] - displ_mean)/displ_std

# non-normalized and normalized car
car = {}
car['year'] = 1985
car['cylinders'] = 4
car['displ'] = 5

norm_car = {}
norm_car['year'] = (1985 - year_mean)/year_std
norm_car['cylinders'] = (4- cylinders_mean)/cylinders_std
norm_car['displ'] = (5 - displ_mean)/displ_std

print('non-normalized knn: ', knn(car, subset, 5))
print('normalized knn: ', knn(norm_car, norm_subset, 5))
```

## Describe and explain the differences:
The non-normalized fifth nearest neighbor prediction is 15.6 mpg, whereas the normalized fifth nearest neighbor prediction is a bit lower at 15 mpg.

The normalized data was with a Z-score. A Z-score is a standardized value that compares the individual point with the mean of the set. This way, the normalized prediction is likely a bit more accurate because you could look at the dataset it came from as a level playing field, in a sense.
